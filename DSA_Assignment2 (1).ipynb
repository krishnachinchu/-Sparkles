```python
# Importing Pandas Library as 'pd'
import pandas as pd
# Importing NumPy Library as 'np'
import numpy as np
# Importing Matplot's Sub Library PyPlot as 'plt'
import matplotlib.pyplot as plt
# Importing seaborn Library as 'sns'
import seaborn as sns
```

# 1. Read the dataset to the python environment.

```python
# Getting data from Local CSV File and copying into cars_data Datafram
cars_data=pd.read_csv('cars_data.csv')
```

```python
# Getting data of dataframe cars_data
cars_data.head()
```

# 2. Check for the null values present in the dataset

```python
#To knowing the sum of null value is pandas dataframe
cars_data.isnull().sum().sum()
```

```python
#To find rows with NaN values in a pandas DataFrame in Python
is_NaN = cars_data.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = cars_data[row_has_NaN]
print(rows_with_NaN)
```

```python
# Checking the null values present in the dataset cars_data using .isna()
cars_data[cars_data.isna()]
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
#Plot countplot for buyer gender 
#set background style as darkgrid using seaborn.
sns.set(style="darkgrid")
#Plot count plot using seaborn.
ax = sns.countplot(x='Buyer Gender',  data=cars_data)
#Display count plot title.
plt.title( "Count plot for Buyer gender",Fontsize=16)
plt.show()
```

# Insights
Here, we plotted data from the cars_data dataframe. We indicated that we wanted to plot variables in this dataset with the code data = cars_data.

We used the x parameter to specify a categorical variable to put on the x-axis. The two possible categories of this variable are male and female.

Notice a few things about the syntax

Calling sns.countplot() like this generates a vertical bar chart, where the categories are along the x-axis, and the length of each bar corresponds to the count of the number of records for that particular category.

# 4. Find the top 5 cars based on their sales price.

```python
#Sort data based on their sales price in decending order.
cars_data.sort_values(by=['Sale Price'], inplace=True,ascending=False)
#Display top five raw with index.
cars_data. reset_index ( drop = True ).head(5)
```

# 5. Find the least 5 cars based on their Resell price.

```python
#Sort data based on their sales price in asecending order.
cars_data.sort_values(by=['Resell Price'], inplace=True)
#Display top five raw with index.
cars_data. reset_index ( drop = True ).head(5)
```
